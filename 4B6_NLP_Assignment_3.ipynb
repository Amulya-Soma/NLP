# **Question 1**

Importing nltk for tokenization

```python
import nltk
nltk.download('punkt')
```

Tokenizing given data into sentences

```python
tokenizer = nltk.data.load('tokenizers/punkt/english.pickle')

paragraph_data = "A paragraph is a series of sentences that are organized and coherent, and are all related to a single topic. Almost every piece of writing you do that is longer than a few sentences should be organized into paragraphs. This is because paragraphs show a reader where the subdivisions of an essay begin and end, and thus help the reader see the organization of the essay and grasp its main points. Paragraphs can contain many different kinds of information. A paragraph could contain a series of brief examples or a single long illustration of a general point. It might describe a place, character, or process; narrate a series of events; compare or contrast two or more things; classify items into categories; or describe causes and effects. Regardless of the kind of information they contain, all paragraphs share certain characteristics. One of the most important of these is a topic sentence."
sentences = tokenizer.tokenize(paragraph_data)
print(sentences) #Printing the tokenized sentences
```

# i) Word2Vec Embeddings

```python
import tensorflow_hub as hub
word_2_vector = hub.KerasLayer("https://tfhub.dev/google/Wiki-words-500/2")
print(word_2_vector(nltk.tokenize.word_tokenize(paragraph_data)))
```

# ii) USE Embeddings

```python
import tensorflow_hub as hub
use_embeddings = hub.KerasLayer("https://tfhub.dev/google/nnlm-en-dim128/2")
print(use_embeddings(sentences))
```

# iii) ELMO Embeddings

Importing Elmo link through tensorflow hub and embedding

```python
import tensorflow_hub as hub
import tensorflow.compat.v1 as tf
tf.disable_eager_execution()
elmo_embed=hub.Module("https://tfhub.dev/google/elmo/3",trainable=True)
embeddings=elmo_embed(
    sentences,
    signature="default",
    as_dict=True)["elmo"]
```

Printing the shape of the embeddings

```python
initial=tf.initialize_all_variables()
sess=tf.Session()
sess.run(initial)
print(sess.run(embeddings[0]))
print("shape=",embeddings[0].shape)
```

# **Question 3**

Similar sentences using sentence Bert

```python
from sentence_transformers import SentenceTransformer
sent_bert_model = SentenceTransformer('bert-base-nli-mean-tokens')
import numpy as np
```

```python
sentence_embeddings = sent_bert_model.encode(sentences)
q1_vec = sent_bert_model.encode(sentences[0])
```

```python
def cosine(u, v):
    return np.dot(u, v) / (np.linalg.norm(u) * np.linalg.norm(v))

for sent in sentences:
  sim = cosine(q1_vec, sent_bert_model.encode([sent])[0])
  if sim>0.6:
    print("Sentence - 1 =",sentences[0],"\n \nSentence - 2=", sent, "\n\nSimilarity = \n", sim,end="\n ==================================\n")
```

# Explain POS Tagging with HMM?


HMM (Hidden Markov Model) is a Stochastic technique for POS tagging. Hidden Markov models are known for their applications to reinforcement learning and temporal pattern recognition such as speech, handwriting, gesture recognition, musical score following, partial discharges, and bioinformatics. Associating each word in a sentence with a proper POS (part of speech) is known as POS tagging or POS annotation. POS tags are also known as word classes, morphological classes, or lexical tags. POS tags give a large amount of information about a word and its neighbors. They are used as an intermediate step for higher-level NLP tasks such as parsing, semantics analysis, translation, and many more, which makes POS tagging a necessary function for advanced NLP applications.


```python
import nltk
from nltk import word_tokenize
nltk.download('averaged_perceptron_tagger')
```

```python
for sents in sentences:
  print(nltk.pos_tag(word_tokenize(sents)))
```
